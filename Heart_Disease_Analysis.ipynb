## Visualise Age Distribution by Gender:

What do you interpret from this plot?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

# Load the dataset
df = pd.read_csv('Cardiovascular_Disease_Dataset.csv')

# Set the size of the plot
plt.figure(figsize=(10, 6))

# Create a histogram plot for age distribution by gender
sns.histplot(x='age', hue='gender', data=df, palette='muted', multiple='stack', bins=15)

# Add title and labels
plt.title('Age Distribution by Gender')
plt.xlabel('Age')
plt.ylabel('Count')

# Add legend with custom labels
plt.legend(title='Gender', labels=['Female', 'Male'])

# Show the plot
plt.show()
```

## Annotating the Box Plot

## Visualization:

What do you interpret from this plot?

```python
chestpain_labels = {0: 'normal', 1: 'ST-T wave abnormality', 2: 'probable or definite'}
counts = df['restingrelectro'].value_counts() 

# Replace index with descriptive labels using chestpain_labels dictionary
counts.index = counts.index.map(chestpain_labels)

# Plotting
ax = counts.plot(kind='bar', color='skyblue', title='Distribution of Resting Electrocardiogram Results')
ax.set_xlabel('Resting Electrocardiogram Results')
ax.set_ylabel('Count')

# Annotating bars with counts
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', color='black')

plt.show()
```

The plot provides a clear visualization of the distribution of resting electrocardiogram results in the dataset. It highlights that while most subjects have normal ECG results, a notable number show ST-T wave abnormalities, and a smaller group exhibits left ventricular hypertrophy. This information can be useful for understanding the prevalence of different heart conditions in the study population and may guide further analysis or interventions
